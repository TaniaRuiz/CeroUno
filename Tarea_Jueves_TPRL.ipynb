###Muestra los países con esperanza de vida femenina mayor al promedio general y sus datos



```python
OCDE5=OCDE
OCDE_vida=OCDE5.set_index("INDICATOR")
expectativa=OCDE_vida.loc["HS_LEB"]
pd.DataFrame(expectativa)
```

```python
esperanza=expectativa.set_index("Inequality")
```

```python
mujeres=esperanza.loc["Women"]
pd.DataFrame(mujeres)
```

```python
mujeres.describe()
```

```python
#variable promedio de vida
Promedio_de_vida=82.24 

total=mujeres.count()
print(total)


```

```python
# creación panda para index

mujer_esperanza=mujeres["Value"]
pd.DataFrame(mujer_esperanza)
```

```python
#parámetro para esperanza de vida
Esperanza_vida_promedio=82.24
total=mujer_esperanza.count()
```

```python
P_V=sum(mujer_esperanza>Esperanza_vida_promedio)
Promedio=P_V/total

print("Existen", P_V,"países con expectativa mayor al promedio", "de un total de", total)

      
```

```python
### Cuales países son:

mujeres_nueva=mujeres.drop(["MEASURE","Unit Code","Flags","Measure","INEQUALITY","LOCATION", "PowerCode Code","Reference Period Code","Flag Codes","Reference Period","PowerCode","Unit"],axis=1)
mujer_esperanza_paises=mujeres_nueva[mujeres_nueva["Value"]>82.24]

#Paises en orden 
Paises_ordeandos=mujer_esperanza_paises.sort_values(by=["Value"],axis=0, ascending=False)
pd.DataFrame(Paises_ordeandos)
```
